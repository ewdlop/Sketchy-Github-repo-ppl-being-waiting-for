```python
# 👻 鬼魂 ASCII 藝術生成器
import time
import random

def haunted_ghost():
    """生成一個會飄動的詭異鬼魂"""
    ghosts = [
        """
        .--.
       /    \\
      |  👻  |
      |      |
      |______|
       ~ ~ ~
        """,
        """
        .--.
       /    \\
      |  😱  |
      |      |
      |______|
      ~ ~ ~ ~
        """,
        """
        .--.
       /    \\
      |  💀  |
      |      |
      |______|
       ~ ~ ~
        """
    ]
    return random.choice(ghosts)

# 讓鬼魂出現！
print("🎃 萬聖節快樂！鬼魂正在出沒... 🎃\n")
for i in range(3):
    print(haunted_ghost())
    print("\n" + "="*30 + "\n")
```

## 🦇 蝙蝠洞穴模擬器

模擬蝙蝠在洞穴中飛行的軌跡！


```python
# 🦇 蝙蝠洞穴
import random

def bat_cave(num_bats=20):
    """生成蝙蝠洞穴的視覺效果"""
    cave = []
    width = 50
    
    print("🕸️" + "=" * width + "🕸️")
    print("  🦇 歡迎來到蝙蝠洞穴 🦇")
    print("🕸️" + "=" * width + "🕸️\n")
    
    for i in range(10):
        line = [' '] * width
        # 隨機放置蝙蝠
        for _ in range(random.randint(0, 3)):
            pos = random.randint(0, width - 1)
            line[pos] = '🦇'
        
        # 隨機放置蜘蛛網
        for _ in range(random.randint(0, 2)):
            pos = random.randint(0, width - 1)
            if line[pos] == ' ':
                line[pos] = '🕸️'
        
        print(''.join(line))
    
    print("\n" + "💀" * 10)

bat_cave()
```

## 👻 鬼屋探險遊戲

一個簡單的文字冒險遊戲！


```python
# 👻 鬼屋探險
import random

class HauntedHouse:
    def __init__(self):
        self.rooms = {
            "入口大廳": ["你看到了一個巨大的蜘蛛網", "地上有奇怪的腳印", "吊燈在搖晃"],
            "廚房": ["鍋裡正在煮著可疑的東西", "櫥櫃自己打開了", "聽到奇怪的笑聲"],
            "臥室": ["床下有東西在移動", "鏡子裡倒影不對勁", "窗簾後面有影子"],
            "地下室": ["牆上有血跡", "聽到鐵鏈聲", "看到一雙發光的眼睛"],
            "閣樓": ["塵土飛揚", "發現一本古老的魔法書", "蝙蝠從窗戶飛進來"]
        }
        
        self.monsters = ["殭屍", "吸血鬼", "狼人", "鬼魂", "女巫", "骷髏"]
    
    def explore(self):
        print("🏚️  歡迎來到詭異鬼屋！🏚️ ")
        print("=" * 50)
        print("\n你是一名勇敢的探險家...")
        print("準備探索這棟被遺棄的鬼屋！\n")
        
        for room_name, events in self.rooms.items():
            print(f"\n🚪 你進入了【{room_name}】...")
            print(f"   {random.choice(events)}")
            
            if random.random() > 0.5:
                monster = random.choice(self.monsters)
                print(f"   ⚠️  突然！一個{monster}出現了！")
                
                if random.random() > 0.3:
                    print(f"   ✅ 你成功逃脫了！")
                else:
                    print(f"   💀 你被{monster}嚇暈了...")
                    print(f"   🏃 你驚慌失措地逃出了鬼屋！")
                    print("\n" + "="*50)
                    print("遊戲結束！")
                    return
        
        print("\n" + "="*50)
        print("🎉 恭喜！你成功探索了整棟鬼屋！")
        print("你是真正的勇者！👻🏆")

# 開始探險
game = HauntedHouse()
game.explore()
```
