## 👻 第一課：幽靈詞向量 (Ghost Word Vectors)

**詞向量 (Word Embeddings)** 是 NLP 中最神秘的概念之一——把文字轉換成數字向量，就像把靈魂封印進容器裡！

### 🔮 概念解釋

想像每個單詞都是一個**幽靈**，它們在高維空間中漂浮。相似的幽靈會聚集在一起：
- 「殭屍」和「吸血鬼」的幽靈很接近（都是不死生物）
- 「南瓜」和「萬聖節」的幽靈也會在附近徘徊
- 「愛」和「恨」的幽靈位置相反，但能量相近

這就是詞向量的魔法！✨


```python
# 👻 幽靈詞向量實作
import numpy as np
import random

class GhostWordVector:
    """幽靈詞向量類別 - 將單詞轉換為向量空間中的幽靈"""
    
    def __init__(self, dimension=3):
        """
        初始化幽靈向量空間
        dimension: 向量維度（3D 比較容易視覺化）
        """
        self.dimension = dimension
        self.ghost_vault = {}  # 存放所有幽靈的墓穴
        
    def summon_ghost(self, word):
        """召喚一個新的幽靈（創建詞向量）"""
        if word not in self.ghost_vault:
            # 為每個單詞創建隨機向量（在真實NLP中會用訓練好的向量）
            vector = np.random.randn(self.dimension)
            # 正規化向量（讓幽靈的能量標準化）
            vector = vector / np.linalg.norm(vector)
            self.ghost_vault[word] = vector
            print(f"👻 成功召喚幽靈：'{word}'")
        else:
            print(f"⚠️  幽靈 '{word}' 已經存在！")
        
        return self.ghost_vault[word]
    
    def ghost_similarity(self, word1, word2):
        """計算兩個幽靈的相似度（餘弦相似度）"""
        if word1 not in self.ghost_vault or word2 not in self.ghost_vault:
            print("❌ 有些幽靈還未被召喚！")
            return None
        
        vec1 = self.ghost_vault[word1]
        vec2 = self.ghost_vault[word2]
        
        # 餘弦相似度：向量點積
        similarity = np.dot(vec1, vec2)
        
        return similarity
    
    def find_haunted_neighbors(self, word, top_n=3):
        """找到與指定幽靈最接近的其他幽靈"""
        if word not in self.ghost_vault:
            print(f"❌ 幽靈 '{word}' 尚未被召喚！")
            return []
        
        target_vector = self.ghost_vault[word]
        similarities = {}
        
        for other_word, other_vector in self.ghost_vault.items():
            if other_word != word:
                similarity = np.dot(target_vector, other_vector)
                similarities[other_word] = similarity
        
        # 排序並返回最相似的
        sorted_ghosts = sorted(similarities.items(), key=lambda x: x[1], reverse=True)
        return sorted_ghosts[:top_n]

# 🎃 實驗開始！
print("🔮 啟動幽靈詞向量實驗...\n")

# 創建幽靈向量空間
haunted_space = GhostWordVector(dimension=5)

# 召喚萬聖節相關的幽靈
halloween_words = [
    "南瓜", "鬼魂", "殭屍", "吸血鬼", "女巫", 
    "蜘蛛", "蝙蝠", "墓地", "骷髏", "魔法",
    "黑貓", "掃帚", "魔藥", "月亮", "黑夜"
]

print("=" * 50)
print("召喚儀式開始...")
print("=" * 50 + "\n")

for word in halloween_words:
    haunted_space.summon_ghost(word)

print("\n" + "=" * 50)
print("所有幽靈已就位！")
print("=" * 50)
```

```python
# 🔍 測試幽靈相似度

print("\n🔮 測試幽靈之間的連結...\n")

test_pairs = [
    ("南瓜", "萬聖節"),
    ("吸血鬼", "殭屍"),
    ("女巫", "魔法"),
    ("蜘蛛", "蝙蝠"),
    ("黑貓", "女巫")
]

# 為了演示，我們手動設定一些"訓練好"的向量
# 在真實場景中，這些向量會透過大量文本訓練得到
halloween_trained_vectors = {
    "南瓜": np.array([0.8, 0.6, 0.1, 0.3, 0.5]),
    "鬼魂": np.array([0.2, 0.9, 0.8, 0.7, 0.1]),
    "殭屍": np.array([0.3, 0.85, 0.7, 0.6, 0.2]),
    "吸血鬼": np.array([0.25, 0.88, 0.75, 0.65, 0.15]),
    "女巫": np.array([0.5, 0.6, 0.7, 0.8, 0.4]),
    "魔法": np.array([0.45, 0.55, 0.65, 0.75, 0.35]),
    "蜘蛛": np.array([0.6, 0.4, 0.5, 0.3, 0.7]),
    "蝙蝠": np.array([0.55, 0.45, 0.48, 0.35, 0.65]),
    "黑貓": np.array([0.48, 0.58, 0.68, 0.78, 0.38]),
}

# 正規化所有向量
for word, vec in halloween_trained_vectors.items():
    halloween_trained_vectors[word] = vec / np.linalg.norm(vec)
    haunted_space.ghost_vault[word] = halloween_trained_vectors[word]

print("=" * 60)
for word1, word2 in test_pairs:
    similarity = haunted_space.ghost_similarity(word1, word2)
    if similarity is not None:
        # 轉換為百分比
        percent = (similarity + 1) * 50  # 餘弦相似度範圍 [-1, 1] 轉為 [0, 100]
        
        if percent > 70:
            emoji = "🔥"
            status = "非常接近"
        elif percent > 50:
            emoji = "✨"
            status = "有些關聯"
        else:
            emoji = "💨"
            status = "相距甚遠"
        
        print(f"{emoji} '{word1}' ↔️ '{word2}': {percent:.1f}% 相似度 ({status})")

print("=" * 60)
```

```python
# 🔮 幽靈鄰居搜尋

print("\n\n👻 找出每個幽靈最親密的夥伴...\n")

search_words = ["女巫", "殭屍", "南瓜"]

for word in search_words:
    print(f"\n🔍 尋找與 '{word}' 最接近的幽靈...")
    neighbors = haunted_space.find_haunted_neighbors(word, top_n=3)
    
    if neighbors:
        print(f"   👻 最接近的幽靈夥伴：")
        for i, (neighbor, similarity) in enumerate(neighbors, 1):
            percent = (similarity + 1) * 50
            bar = "█" * int(percent / 5)
            print(f"   {i}. {neighbor} - {bar} {percent:.1f}%")
    
    print("   " + "-" * 40)
```

## 🧟 第二課：殭屍神經網絡 (Zombie Neural Networks)

**神經網絡**就像殭屍大軍——每個神經元都是一個殭屍，它們：
- 🧟 接收信號（聞到人腦的味道）
- 🔄 處理信息（決定往哪個方向走）
- 📤 傳遞輸出（發出殭屍吼聲通知其他殭屍）

### 🧠 為什麼叫「殭屍」網絡？

1. **層層推進**：就像殭屍潮，一層層神經元依序處理資訊
2. **集體智慧**：單個殭屍（神經元）很笨，但成群結隊就能完成複雜任務
3. **不死特性**：訓練好的神經網絡可以一直使用，永不疲倦！
4. **反向傳播**：就像殭屍病毒往回傳染，錯誤信號往回修正權重


```python
# 🧟 殭屍神經網絡實作

import numpy as np

class ZombieNeuron:
    """單個殭屍神經元"""
    
    def __init__(self, name, threshold=0.5):
        self.name = name
        self.threshold = threshold  # 殭屍的「覺醒閾值」
        self.is_active = False
        
    def receive_signal(self, signal_strength):
        """接收信號並決定是否被激活（殭屍甦醒）"""
        if signal_strength >= self.threshold:
            self.is_active = True
            return 1.0  # 完全激活
        else:
            self.is_active = False
            return 0.0  # 沉睡中
    
    def __str__(self):
        status = "🧟 甦醒" if self.is_active else "😴 沉睡"
        return f"殭屍 {self.name}: {status}"


class ZombieNeuralNetwork:
    """殭屍神經網絡 - 簡單的前饋網絡"""
    
    def __init__(self):
        print("⚰️ 正在組建殭屍大軍...\n")
        
        # 創建三層殭屍網絡
        self.input_layer = [
            ZombieNeuron("輸入殭屍_1", 0.3),
            ZombieNeuron("輸入殭屍_2", 0.3),
            ZombieNeuron("輸入殭屍_3", 0.3)
        ]
        
        self.hidden_layer = [
            ZombieNeuron("隱藏殭屍_A", 0.5),
            ZombieNeuron("隱藏殭屍_B", 0.5)
        ]
        
        self.output_layer = [
            ZombieNeuron("輸出殭屍", 0.4)
        ]
        
        # 連接權重（殭屍之間的聯繫強度）
        self.weights_input_hidden = np.random.randn(3, 2) * 0.5
        self.weights_hidden_output = np.random.randn(2, 1) * 0.5
        
        print("✅ 殭屍大軍集結完畢！")
        print(f"   - 輸入層：{len(self.input_layer)} 個殭屍")
        print(f"   - 隱藏層：{len(self.hidden_layer)} 個殭屍")
        print(f"   - 輸出層：{len(self.output_layer)} 個殭屍")
        print()
    
    def forward_propagation(self, input_signals):
        """前向傳播 - 殭屍潮湧向目標！"""
        print("🧟‍♂️🧟‍♀️🧟 殭屍潮開始行動...\n")
        
        # 第一層：輸入層激活
        print("📥 第一波殭屍甦醒：")
        input_activations = []
        for i, zombie in enumerate(self.input_layer):
            activation = zombie.receive_signal(input_signals[i])
            input_activations.append(activation)
            print(f"   {zombie}")
        
        input_activations = np.array(input_activations)
        
        # 第二層：隱藏層激活
        print("\n⚡ 信號傳遞到隱藏層...")
        hidden_signals = np.dot(input_activations, self.weights_input_hidden)
        hidden_activations = []
        for i, zombie in enumerate(self.hidden_layer):
            activation = zombie.receive_signal(hidden_signals[i])
            hidden_activations.append(activation)
            print(f"   {zombie}")
        
        hidden_activations = np.array(hidden_activations)
        
        # 第三層：輸出層激活
        print("\n📤 最終輸出...")
        output_signal = np.dot(hidden_activations, self.weights_hidden_output)[0]
        output_activation = self.output_layer[0].receive_signal(output_signal)
        print(f"   {self.output_layer[0]}")
        
        print(f"\n🎯 最終輸出強度：{output_activation:.2f}")
        
        return output_activation

# 🎃 創建並測試殭屍網絡
print("="*60)
print(" " * 15 + "🧟 殭屍神經網絡實驗 🧟")
print("="*60 + "\n")

zombie_net = ZombieNeuralNetwork()

print("\n" + "="*60)
print("測試案例：偵測是否為萬聖節")
print("="*60 + "\n")

# 輸入特徵：[有南瓜(0.8), 是晚上(0.9), 有糖果(0.7)]
test_input = [0.8, 0.9, 0.7]
print(f"輸入特徵：")
print(f"  - 有南瓜：{test_input[0]}")
print(f"  - 是晚上：{test_input[1]}")
print(f"  - 有糖果：{test_input[2]}")
print()

result = zombie_net.forward_propagation(test_input)

if result > 0.5:
    print("\n🎃 結論：這是萬聖節！殭屍們很興奮！")
else:
    print("\n😴 結論：不是萬聖節，殭屍們繼續沉睡...")

print("\n" + "="*60)
```

```python
# 🧛 吸血鬼注意力機制實作

import numpy as np

class VampireAttention:
    """吸血鬼注意力機制 - 選擇性吸取重要信息"""
    
    def __init__(self, name="德古拉伯爵"):
        self.name = name
        self.victims = []  # 被注意的詞彙（受害者）
        self.attention_scores = []  # 注意力分數
        
    def calculate_attention(self, query, keys):
        """
        計算注意力分數
        query: 吸血鬼當前的目標（想要理解的詞）
        keys: 所有可能的受害者（句子中的所有詞）
        """
        print(f"🧛 {self.name} 開始尋找獵物...\n")
        
        # 計算相似度分數（點積）
        scores = []
        for i, key in enumerate(keys):
            # 簡單的相似度計算
            score = np.dot(query, key)
            scores.append(score)
        
        scores = np.array(scores)
        
        # Softmax - 轉換為概率分布（吸血鬼的「飢餓指數」）
        exp_scores = np.exp(scores - np.max(scores))  # 數值穩定性
        attention_weights = exp_scores / np.sum(exp_scores)
        
        self.attention_scores = attention_weights
        
        return attention_weights
    
    def visualize_attention(self, words, attention_weights):
        """視覺化注意力分布"""
        print("🎯 吸血鬼的注意力分布（誰最吸引他）：\n")
        
        max_weight = np.max(attention_weights)
        
        for word, weight in zip(words, attention_weights):
            # 計算視覺化長度
            bar_length = int(weight / max_weight * 30)
            bar = "🩸" * bar_length
            percentage = weight * 100
            
            # 判斷注意力等級
            if percentage > 30:
                emoji = "🧛"
                status = "主要目標"
            elif percentage > 10:
                emoji = "👀"
                status = "次要目標"
            else:
                emoji = "💨"
                status = "忽略"
            
            print(f"{emoji} '{word}': {bar} {percentage:.1f}% ({status})")
    
    def extract_blood(self, values, attention_weights):
        """
        吸取信息（加權求和）
        values: 每個詞的實際信息內容
        attention_weights: 注意力權重
        """
        # 加權平均
        extracted_info = np.zeros_like(values[0])
        for value, weight in zip(values, attention_weights):
            extracted_info += weight * value
        
        return extracted_info


# 🎃 實驗：吸血鬼理解句子

print("="*70)
print(" " * 20 + "🧛 吸血鬼注意力實驗 🧛")
print("="*70 + "\n")

# 模擬一個句子："萬聖節的夜晚，南瓜燈照亮了黑暗"
sentence = ["萬聖節", "的", "夜晚", "南瓜燈", "照亮", "了", "黑暗"]

# 為每個詞創建簡單的向量表示（實際應用中會用訓練好的詞向量）
word_vectors = {
    "萬聖節": np.array([0.9, 0.8, 0.2]),
    "的": np.array([0.1, 0.1, 0.1]),
    "夜晚": np.array([0.2, 0.1, 0.9]),
    "南瓜燈": np.array([0.85, 0.75, 0.3]),
    "照亮": np.array([0.7, 0.3, 0.6]),
    "了": np.array([0.1, 0.1, 0.1]),
    "黑暗": np.array([0.1, 0.2, 0.95])
}

print("📖 原句：「" + " ".join(sentence) + "」\n")
print("-"*70 + "\n")

# 創建吸血鬼
dracula = VampireAttention("德古拉伯爵")

# 假設我們想理解「萬聖節」這個詞的上下文
query_word = "萬聖節"
query_vector = word_vectors[query_word]

print(f"🎯 查詢詞（吸血鬼的初始目標）：'{query_word}'\n")

# 準備所有詞的向量
keys = [word_vectors[w] for w in sentence]
values = keys  # 在這個簡化例子中，keys 和 values 相同

# 計算注意力
attention_weights = dracula.calculate_attention(query_vector, keys)

# 視覺化
dracula.visualize_attention(sentence, attention_weights)

print("\n" + "-"*70)

# 吸取信息
extracted_info = dracula.extract_blood(values, attention_weights)

print("\n🩸 吸血鬼吸取的信息（加權向量）：")
print(f"   {extracted_info}")

print("\n💡 解釋：")
print("   - 注意力最高的詞會被重點關注")
print("   - 「的」、「了」等虛詞注意力較低")
print("   - 相關名詞（南瓜燈）獲得較高注意力")
print("   - 最終輸出是所有詞的加權組合")

print("\n" + "="*70)
```

```python
# 🕷️ 蜘蛛網語言模型

from collections import defaultdict
import random

class SpiderWebLanguageModel:
    """蜘蛛網 N-gram 語言模型"""
    
    def __init__(self, n=2):
        """
        n: N-gram 的 N（蜘蛛的記憶長度）
        n=2: bigram（記住前1個詞）
        n=3: trigram（記住前2個詞）
        """
        self.n = n
        self.web = defaultdict(list)  # 蜘蛛網結構
        self.start_words = []  # 網的入口點
        
    def build_web(self, texts):
        """編織蜘蛛網（訓練模型）"""
        print(f"🕷️ 蜘蛛開始編織 {self.n}-gram 網...\n")
        
        for text in texts:
            words = text.split()
            
            # 記錄開頭詞
            if len(words) >= self.n:
                self.start_words.append(tuple(words[:self.n-1]))
            
            # 建立 N-gram 連接
            for i in range(len(words) - self.n + 1):
                # 取前 n-1 個詞作為「上下文」
                context = tuple(words[i:i+self.n-1])
                next_word = words[i+self.n-1]
                
                # 在蜘蛛網中建立連接
                self.web[context].append(next_word)
        
        print(f"✅ 蜘蛛網編織完成！")
        print(f"   - 網絡節點數：{len(self.web)}")
        print(f"   - 入口點數：{len(self.start_words)}")
        
    def predict_next(self, context):
        """預測下一個詞（蜘蛛爬行）"""
        context_tuple = tuple(context.split())
        
        if context_tuple in self.web:
            # 隨機選擇一個可能的下一個詞
            next_words = self.web[context_tuple]
            return random.choice(next_words)
        else:
            return None
    
    def generate_sentence(self, max_words=15):
        """生成句子（蜘蛛在網上爬行）"""
        if not self.start_words:
            return "🕸️ 網還沒編好..."
        
        # 隨機選擇一個起點
        current_words = list(random.choice(self.start_words))
        
        for _ in range(max_words):
            # 取最後 n-1 個詞作為上下文
            context = tuple(current_words[-(self.n-1):])
            
            # 預測下一個詞
            if context in self.web:
                next_word = random.choice(self.web[context])
                current_words.append(next_word)
            else:
                # 網斷了，停止生成
                break
        
        return " ".join(current_words)
    
    def show_web_structure(self, sample_size=5):
        """展示蜘蛛網的部分結構"""
        print("\n🕸️ 蜘蛛網結構樣本（部分連接）：\n")
        
        items = list(self.web.items())[:sample_size]
        
        for context, next_words in items:
            context_str = " ".join(context)
            print(f"📍 '{context_str}' → 可能接：")
            
            # 統計每個詞出現的次數
            word_counts = {}
            for word in next_words:
                word_counts[word] = word_counts.get(word, 0) + 1
            
            for word, count in word_counts.items():
                probability = count / len(next_words) * 100
                print(f"      🕷️ '{word}' (機率: {probability:.1f}%)")
            print()


# 🎃 實驗：訓練萬聖節語言模型

print("="*70)
print(" " * 18 + "🕷️ 蜘蛛網語言模型實驗 🕷️")
print("="*70 + "\n")

# 訓練數據：萬聖節相關句子
halloween_corpus = [
    "萬聖節的夜晚 南瓜燈照亮黑暗",
    "萬聖節的派對 孩子們穿著可怕的服裝",
    "南瓜燈在 萬聖節 是重要的象徵",
    "黑暗的夜晚 吸血鬼出沒在街道上",
    "可怕的服裝 是萬聖節的傳統",
    "萬聖節的糖果 讓孩子們興奮不已",
    "鬼魂在萬聖節的夜晚出現",
    "南瓜燈的光芒在黑暗中閃爍",
    "孩子們在萬聖節收集糖果",
    "可怕的裝飾 布滿了整個街道"
]

# 創建並訓練模型
spider = SpiderWebLanguageModel(n=2)  # Bigram 模型
spider.build_web(halloween_corpus)

# 展示網絡結構
spider.show_web_structure(sample_size=8)

print("-"*70)
print("\n🎲 讓蜘蛛生成一些萬聖節句子：\n")

for i in range(5):
    sentence = spider.generate_sentence(max_words=12)
    print(f"{i+1}. 🕸️ {sentence}")

print("\n" + "="*70)

print("\n💡 語言模型的應用：")
print("   ✅ 自動補全（手機輸入法）")
print("   ✅ 機器翻譯")
print("   ✅ 文本生成（ChatGPT、GPT-4）")
print("   ✅ 語音識別")
print("   ✅ 拼寫檢查")
```

## 🎃 第五課：巫師的詛咒分類器 (Witch's Curse Classifier)

**文本分類**是 NLP 中最實用的應用之一！就像女巫區分不同的詛咒類型。

### 🔮 應用場景
- 📧 垃圾郵件偵測（惡意詛咒 vs 正常郵件）
- 💬 情感分析（快樂詛咒 vs 悲傷詛咒）
- 🏷️ 新聞分類（政治詛咒、體育詛咒、娛樂詛咒）
- 🤖 意圖識別（聊天機器人理解用戶想要什麼）


```python
# 🧙‍♀️ 女巫的詛咒分類器

from collections import Counter
import re

class WitchCurseClassifier:
    """女巫的詛咒分類器 - 簡單的樸素貝葉斯分類器"""
    
    def __init__(self):
        self.word_counts = {}  # 每個類別的詞頻
        self.class_counts = {}  # 每個類別的文檔數
        self.vocabulary = set()  # 詞彙表
        
    def preprocess(self, text):
        """預處理文本（女巫的咒語準備）"""
        # 轉小寫、分詞
        words = text.lower().split()
        return words
    
    def train(self, texts, labels):
        """訓練分類器（女巫學習詛咒）"""
        print("🔮 女巫開始學習詛咒的奧秘...\n")
        
        for text, label in zip(texts, labels):
            words = self.preprocess(text)
            
            # 更新詞彙表
            self.vocabulary.update(words)
            
            # 初始化類別計數
            if label not in self.word_counts:
                self.word_counts[label] = Counter()
                self.class_counts[label] = 0
            
            # 統計詞頻
            self.word_counts[label].update(words)
            self.class_counts[label] += 1
        
        print("✅ 學習完成！")
        print(f"   - 詛咒類型數：{len(self.class_counts)}")
        print(f"   - 詞彙量：{len(self.vocabulary)}")
        print(f"   - 訓練樣本數：{sum(self.class_counts.values())}")
        
        for curse_type, count in self.class_counts.items():
            print(f"     • {curse_type}：{count} 個樣本")
    
    def predict(self, text):
        """預測文本類別（識別詛咒類型）"""
        words = self.preprocess(text)
        
        scores = {}
        total_docs = sum(self.class_counts.values())
        
        for label in self.class_counts:
            # 計算先驗概率 P(class)
            prior = self.class_counts[label] / total_docs
            
            # 計算似然概率 P(words|class)
            likelihood = 1.0
            total_words = sum(self.word_counts[label].values())
            vocab_size = len(self.vocabulary)
            
            for word in words:
                # Laplace 平滑
                word_count = self.word_counts[label].get(word, 0)
                word_prob = (word_count + 1) / (total_words + vocab_size)
                likelihood *= word_prob
            
            # 後驗概率（取對數避免下溢）
            scores[label] = prior * likelihood
        
        # 返回分數最高的類別
        return max(scores.items(), key=lambda x: x[1])
    
    def show_top_words(self, label, top_n=5):
        """顯示某類詛咒的特徵詞"""
        if label not in self.word_counts:
            return
        
        top_words = self.word_counts[label].most_common(top_n)
        print(f"\n🔮 '{label}' 詛咒的特徵詞：")
        for word, count in top_words:
            print(f"   ✨ '{word}': {count} 次")


# 🎃 實驗：訓練詛咒分類器

print("="*70)
print(" " * 16 + "🧙‍♀️ 女巫的詛咒分類器實驗 🧙‍♀️")
print("="*70 + "\n")

# 訓練數據：不同類型的詛咒
curse_texts = [
    # 恐怖詛咒
    "黑暗 鬼魂 恐怖 嚇人 可怕 驚悚 死亡 骷髏",
    "吸血鬼 殭屍 怪物 邪惡 恐怖 血腥 墓地",
    "尖叫 恐怖 黑暗 噩夢 恐懼 顫抖",
    
    # 歡樂詛咒
    "糖果 派對 歡笑 快樂 慶祝 遊戲 朋友 開心",
    "禮物 音樂 舞蹈 歡樂 慶典 美食 快樂",
    "歡笑 玩耍 慶祝 派對 開心 愉快",
    
    # 魔法詛咒
    "魔法 咒語 魔杖 變身 飛行 魔藥 法術 神秘",
    "魔法書 咒語 巫師 魔力 變化 法術 神奇",
    "魔法 魔藥 咒語 魔力 變身 神秘 施法"
]

curse_labels = [
    "恐怖", "恐怖", "恐怖",
    "歡樂", "歡樂", "歡樂",
    "魔法", "魔法", "魔法"
]

# 創建並訓練分類器
witch = WitchCurseClassifier()
witch.train(curse_texts, curse_labels)

# 顯示每類詛咒的特徵詞
for label in set(curse_labels):
    witch.show_top_words(label, top_n=5)

print("\n" + "="*70)
print("\n🔮 測試女巫的分類能力：\n")

# 測試案例
test_cases = [
    "這裡有很多鬼魂和骷髏 非常恐怖",
    "我們在派對上玩遊戲 吃糖果 很開心",
    "女巫用魔杖施展了神秘的咒語",
    "黑暗的墓地裡傳來尖叫聲",
    "朋友們一起慶祝 歡笑不停"
]

for i, test_text in enumerate(test_cases, 1):
    predicted_label, score = witch.predict(test_text)
    
    # 選擇 emoji
    emoji_map = {
        "恐怖": "👻",
        "歡樂": "🎉",
        "魔法": "🔮"
    }
    emoji = emoji_map.get(predicted_label, "❓")
    
    print(f"{i}. 測試文本：「{test_text}」")
    print(f"   {emoji} 預測：{predicted_label} (信心度: {score:.6f})")
    print()

print("="*70)
```
